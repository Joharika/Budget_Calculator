```python
# Budget Calculator
# Initialize a dictionary to hold the budget categories and their amounts
budget = {
    "income": [],
    "expenses": []
}
# Function to add an income
def add_income(amount, description):
    budget['income'].append({"amount": amount, "description": description})
    print(f"Income of ${amount} added: {description}")
# Function to add an expense
def add_expense(amount, description):
    budget['expenses'].append({"amount": amount, "description": description})
    print(f"Expense of ${amount} added: {description}")
# Function to calculate total income
def total_income():
    return sum(item["amount"] for item in budget['income'])
# Function to calculate total expenses
def total_expenses():
    return sum(item["amount"] for item in budget['expenses'])
# Function to calculate the balance
def calculate_balance():
    return total_income() - total_expenses()
# Function to display the budget summary
def display_summary():
    print("\nBudget Summary")
    print("------------------")
    print("Income:")
    for income in budget['income']:
        print(f"  {income['description']}: ${income['amount']}")
    print("Total Income: $", total_income())
    print("\nExpenses:")
    for expense in budget['expenses']:
        print(f"  {expense['description']}: ${expense['amount']}")
    print("Total Expenses: $", total_expenses())
    balance = calculate_balance()
    print("\nBalance: $", balance)
    if balance > 0:
        print(f"Great job! You have a positive balance.Your balance amount is {balance} ðŸ˜Š")
    elif balance == 0:
        print("Your balance is zero. Consider reducing your expenses. ðŸ˜Ÿ")
    else:
        print("Oops! You have a negative balance. Consider reviewing your expenses. ðŸ˜Ÿ")
# Main program loop
while True:
    print("\nBudget Calculator")
    print("1. Add Income")
    print("2. Add Expense")
    print("3. View Summary")
    print("4. Exit")
    choice = input("Choose an option (1-4): ")
    if choice == '1':
        amount = float(input("Enter income amount: "))
        description = input("Enter income description: ")
        add_income(amount, description)
    elif choice == '2':
        amount = float(input("Enter expense amount: "))
        description = input("Enter expense description: ")
        add_expense(amount, description)
    elif choice == '3':
        display_summary()
    elif choice == '4':
        print("Exiting Budget Calculator. Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option.")
```
